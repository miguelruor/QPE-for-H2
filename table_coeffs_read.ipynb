```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("table_coeffs.html", "r") as f:
    soup = BeautifulSoup(f.read(), "html.parser")
```

```python
rows = soup.find_all("tr")
with open("table_coeffs.csv", "w") as f:
    f.write("nuclear_dist,I,Z0,Z1,Z0Z1,X0X1,Y0Y1,t0,ordering,trotter_error\n")
    for row in rows[1:]:
        f.write(",".join(td.get_text(strip=True) for td in row.find_all("td")) + "\n")
```
